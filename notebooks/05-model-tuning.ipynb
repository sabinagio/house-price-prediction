```python
from pycaret.regression import *
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/train.csv')
```

```python
df['TotalBaths'] = df['FullBath'] + (0.5 * df['HalfBath'])
df['GarageFinish_Unf'] = df['GarageFinish'] == 'Unf'
```

```python
top = ['OverallQual',
 'GrLivArea',
 'GarageCars',
 'YearBuilt',
 'GarageFinish_Unf',
 'CentralAir',
 'TotalBaths',
 'MSZoning',]
```

```python
from pycaret.regression import *
experiment = setup(data=df[top + ['SalePrice']], target='SalePrice', session_id=123, transform_target=True, keep_features=top, fold=7) 
```

```python
compare_models()
```

```python
gbr = create_model('gbr')
tuned_gbr = tune_model(gbr)
```

```python
test_df = pd.read_csv('../data/raw/test.csv')
test_df['TotalBaths'] = test_df['FullBath'] + (0.5 * test_df['HalfBath'])
test_df['GarageFinish_Unf'] = test_df['GarageFinish'] == 'Unf'
```

```python
predictions = predict_model(estimator=gbr, data=test_df)[['Id', 'prediction_label']].rename({'prediction_label': 'SalePrice'}, axis=1)
predictions.to_csv("../models/submissions/submission2.csv", index=False)
```
